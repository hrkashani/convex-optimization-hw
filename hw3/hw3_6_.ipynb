```python
import cvxpy as cvx
import numpy as np
```

```python
A = np.matrix('1 2 0 1;0 0 3 1; 0 3 1 1; 2 1 2 5; 1 0 3 2')
cmax = np.matrix('100; 100; 100; 100; 100')
p = np.matrix('3; 2; 7; 6')
pdisc = np.matrix('2; 1; 4; 2')
q = np.matrix('4; 10; 5; 10')
```

```python
x = cvx.Variable(4)
x = cvx.reshape(x,(4,1))
t1 = cvx.multiply(p, x)
t2 = cvx.multiply(p, q) + cvx.multiply(pdisc, x-q)
obj = cvx.Maximize(cvx.sum(cvx.minimum(t1, t2)))
cons = [x >= 0, A*x <= cmax]
prob = cvx.Problem(obj, cons)
prob.solve()
r = cvx.minimum(t1, t2).value
sum_r = sum(r)
avgPrice = r / x.value
print("the optimal activity levels is :")
print (x.value,"\n")
print("the revenue generated by each one is :")
print (r,"\n")
print("the total revenue generated by the optimal solution is :")
print (sum_r,"\n")
print("the average price per unit for each activity level is :")
print (avgPrice,"\n")
```
